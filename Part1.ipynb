7 DAYS OF DATA ANALYTICS WITH PYTHON!



```python
import pandas as pd
import numpy as np
```

creating a DataFrame with null values

```python
num=pd.DataFrame({'LETTERS':['TEN','TWENTY','THIRTY','FOURTY','HUNDRED'],'NUMERICS':[10,20,30,40,100],'INCREMENT':[11,21,31,41,101],'DECREMENT':[9,np.nan,29,39,np.nan]},index=[1,2,3,4,5])
```

display the dataframe

```python
num
```

Displaying Null values

```python
num.isnull()
```

taking the total sum of how many null values are there

```python
num.isnull().sum()
```

```python
print(num.isnull().sum())
```

Handling missing values

1.REMOVE NULLS

by row wise

```python
num_cleaned=num.dropna()
num_cleaned
```

by column wise

```python
num_cleancol=num.dropna(axis=1)
num_cleancol
```

see the difference between using dropna() rowwise and column wise?

```python
print("original datasetðŸŒ•:\n",num)
print("\nAfter removing rows with null valuesðŸŒ’:\n", num_cleaned)
print("\nAfter removing columns with null valuesðŸŒ˜:\n",num_cleancol)
```

2. Filling columns with a value

```python
num.fillna('100')
```

3. forward or backward fill

```python
num.ffill() # forward fill
```

```python
numfill=num.bfill() # backward fill
numfill
```

A quick analyzing after cleaning

```python
print(num.describe())
```

```python
num['LETTERS'].value_counts()
```

```python
num.compare(numfill)
```
