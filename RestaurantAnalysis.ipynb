LEVEL 1 -TASK 1

Determine the top three most common cuisines in the dataset.

```python
common = df['Cuisines'].value_counts().head(3)
common
```

Calculate the percentage of restaurants that serve each of the top cuisines.

```python
restaurants = len(df)
per = (common / restaurants) * 100
per.round(2)
```

TASK 2- City Analysis

Identify the city with the highest number of restaurants in the dataset.

```python
city = df['City'].value_counts()
print(city.head(1))
```

Calculate the average rating for restaurants in each city.

```python
average = df.groupby('City')['Aggregate rating'].mean()

print(average)
```

Determine the city with the highest average rating.

```python
toppest_city = average.sort_values(ascending=False).head(1)

print(toppest_city)
```

TASK 3 - Price Range Distribution

Create a histogram or bar chart to visualize the distribution of price ranges among the restaurants.

```python
df['Price range'].value_counts().sort_index().plot(kind='bar')

plt.title("Price Ranges distribution of restaurants")
plt.xlabel("Price Range")
plt.ylabel("No of Restaurants")
plt.show()
```

Calculate the percentage of restaurants in each price range category.

```python
count = df['Price range'].value_counts()
total = len(df)
percen = (count / total) * 100
print(percen.round(2))
```

TASK 4 - Online Delivery 

Determine the percentage of restaurants that offer online delivery.

```python
count = df['Has Online delivery'].value_counts()
percentage = (count / len(df)) * 100
print(percentage.round(2))
```

Compare the average ratings of restaurants with and without online delivery.

```python
average_ratings = df.groupby('Has Online delivery')['Aggregate rating'].mean()
print("Average Ratings Comparison:")
print(average_ratings)
```
