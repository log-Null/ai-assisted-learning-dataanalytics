ðŸ§¹ SECTION 2: HANDLE MISSING DATA (CLEANING METHODS)


1. Drop columns that contain null values.

2. Check and print how the shape of the DataFrame changed after dropping.

3. Replace all missing values 
- RATING with the median of that column.
- RELEASE YEAR with the mean of that column (rounded).


noteðŸ’¡:df['column'] = df['column'].fillna(df['column'].median()) â†’ fill nulls in one column

```python
show_duplicate
```

```python
show_duplicate['RATING'] = show_duplicate['RATING'].fillna(show_duplicate['RATING'].median())
show_duplicate
```

```python
show_duplicate['RELEASE YEAR']=show_duplicate['RELEASE YEAR'].fillna(show_duplicate['RELEASE YEAR'].mean()).astype(int)
show_duplicate
```

4. Perform a forward fill and backward fill, then compare the difference.

```python
forward= show_duplicate.ffill()
forward
```

```python
backward = show_duplicate.bfill()
backward
```

```python
forward.compare(backward)
```
