A BAR CHART PLOTTING FOR NUMBER OF MOVIES PER GENRE

```python

genre_counts = df['MOVIE_GENRE'].value_counts()
plt.bar(genre_counts.index, genre_counts.values, color='YELLOW')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.title('Number of Movies per Genre')
plt.tight_layout()   
plt.show()
```
